```
"""
Detecting Cyber Threats Using Anomaly Detection in Network Traffic
- Uses machine learning to identify anomalous network behavior
- Processes and analyzes network traffic data
- Implements multiple anomaly detection algorithms
- Visualizes results and detected anomalies
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")

# 1. Data Loading and Preprocessing Functions

def load_network_data(file_path):
    """
    Load network traffic data from CSV file
    Expected columns might include: timestamp, src_ip, dst_ip, protocol,
    bytes_sent, bytes_received, duration, etc.
    """
    print(f"Loading data from {file_path}...")
    # For demonstration - replace with actual file path in production
    try:
        df = pd.read_csv(file_path)
        print(f"Successfully loaded data with {df.shape[0]} records and {df.shape[1]} features")
        return df
    except Exception as e:
        print(f"Error loading data: {e}")
        # Sample data for demonstration purposes
        print("Creating sample demonstration data...")

        # Create synthetic data for demonstration
        np.random.seed(42)
        n_samples = 10000

        # Normal traffic patterns (majority of data)
        normal_data = pd.DataFrame({
            'timestamp': pd.date_range(start='2023-01-01', periods=n_samples, freq='1min'),
            'protocol': np.random.choice(['TCP', 'UDP', 'ICMP'], size=n_samples, p=[0.7, 0.25, 0.05]),
            'src_port': np.random.randint(1024, 65535, n_samples),
            'dst_port': np.random.choice([80, 443, 22, 53, 3389] + list(range(1024, 10000)), n_samples),
            'bytes_sent': np.random.normal(1500, 500, n_samples),
            'bytes_received': np.random.normal(3000, 1000, n_samples),
            'duration': np.random.exponential(3, n_samples),
            'packet_count': np.random.poisson(20, n_samples)
        })

        # Add some anomalies (5% of data)
        n_anomalies = int(n_samples * 0.05)
        anomaly_indices = np.random.choice(n_samples, n_anomalies, replace=False)

        # Different types of anomalies
        # 1. Port scanning (high port count, low bytes)
        port_scan_idx = anomaly_indices[:int(n_anomalies/3)]
        normal_data.loc[port_scan_idx, 'dst_port'] = [np.random.randint(1, 1000) for _ in range(len(port_scan_idx))]
        normal_data.loc[port_scan_idx, 'bytes_sent'] = np.random.normal(100, 50, len(port_scan_idx))
        normal_data.loc[port_scan_idx, 'bytes_received'] = np.random.normal(100, 50, len(port_scan_idx))
        normal_data.loc[port_scan_idx, 'packet_count'] = np.random.poisson(200, len(port_scan_idx))
        normal_data.loc[port_scan_idx, 'duration'] = np.random.uniform(0.01, 0.1, len(port_scan_idx))

        # 2. Data exfiltration (high outbound bytes)
        exfil_idx = anomaly_indices[int(n_anomalies/3):int(2*n_anomalies/3)]
        normal_data.loc[exfil_idx, 'bytes_sent'] = np.random.normal(50000, 10000, len(exfil_idx))
        normal_data.loc[exfil_idx, 'duration'] = np.random.normal(10, 2, len(exfil_idx))

        # 3. DoS attack (high packet count, long duration)
        dos_idx = anomaly_indices[int(2*n_anomalies/3):]
        normal_data.loc[dos_idx, 'packet_count'] = np.random.poisson(500, len(dos_idx))
        normal_data.loc[dos_idx, 'duration'] = np.random.normal(30, 5, len(dos_idx))
        normal_data.loc[dos_idx, 'bytes_received'] = np.random.normal(100, 30, len(dos_idx))

        print(f"Created sample data with {normal_data.shape[0]} records and {normal_data.shape[1]} features")
        return normal_data

def preprocess_data(df):
    """
    Preprocess network traffic data:
    - Handle missing values
    - Convert categorical features
    - Feature engineering
    - Normalization
    """
    print("Preprocessing data...")

    # Make a copy to avoid modifying the original
    df_processed = df.copy()

    # 1. Handle missing values
    for col in df_processed.columns:
        if df_processed[col].dtype == np.number:
            df_processed[col].fillna(df_processed[col].median(), inplace=True)
        else:
            df_processed[col].fillna(df_processed[col].mode()[0], inplace=True)

    # 2. Convert timestamp to datetime if it's not already
    if 'timestamp' in df_processed.columns and df_processed['timestamp'].dtype != 'datetime64[ns]':
        df_processed['timestamp'] = pd.to_datetime(df_processed['timestamp'])

        # Extract time-based features
        df_processed['hour'] = df_processed['timestamp'].dt.hour
        df_processed['day_of_week'] = df_processed['timestamp'].dt.dayofweek
        df_processed['is_weekend'] = df_processed['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

    # 3. Convert categorical features using one-hot encoding
    if 'protocol' in df_processed.columns:
        df_processed = pd.get_dummies(df_processed, columns=['protocol'], prefix='proto')

    # 4. Feature engineering
    # Bytes per packet ratio (if we have both features)
    if 'bytes_sent' in df_processed.columns and 'packet_count' in df_processed.columns:
        df_processed['bytes_per_packet'] = df_processed['bytes_sent'] / df_processed['packet_count'].replace(0, 1)

    # Bytes ratio (sent/received)
    if 'bytes_sent' in df_processed.columns and 'bytes_received' in df_processed.columns:
        df_processed['bytes_ratio'] = df_processed['bytes_sent'] / df_processed['bytes_received'].replace(0, 1)

    # Packets per second
    if 'packet_count' in df_processed.columns and 'duration' in df_processed.columns:
        df_processed['packets_per_second'] = df_processed['packet_count'] / df_processed['duration'].replace(0, 0.1)

    # 5. Select only numeric columns for model training
    numeric_cols = df_processed.select_dtypes(include=[np.number]).columns.tolist()

    # Remove timestamp column if it exists in numeric_cols
    if 'timestamp' in numeric_cols:
        numeric_cols.remove('timestamp')

    # 6. Scale the numeric features
    scaler = StandardScaler()
    df_processed[numeric_cols] = scaler.fit_transform(df_processed[numeric_cols])

    print(f"Preprocessing complete. Features used: {numeric_cols}")
    return df_processed, numeric_cols, scaler


# 2. Anomaly Detection Models

def isolation_forest_detection(df, features, contamination=0.05):
    """
    Apply Isolation Forest algorithm for anomaly detection
    """
    print("\nRunning Isolation Forest anomaly detection...")
    model = IsolationForest(contamination=contamination, random_state=42)
    df['isolation_forest_score'] = model.fit_predict(df[features])
    # Convert predictions: -1 for anomalies, 1 for normal points
    df['isolation_forest_anomaly'] = df['isolation_forest_score'].apply(lambda x: 1 if x == -1 else 0)

    anomaly_count = df['isolation_forest_anomaly'].sum()
    print(f"Isolation Forest detected {anomaly_count} anomalies ({anomaly_count/len(df)*100:.2f}%)")
    return df, model

def lof_detection(df, features, contamination=0.05):
    """
    Apply Local Outlier Factor (LOF) for anomaly detection
    """
    print("\nRunning Local Outlier Factor anomaly detection...")
    model = LocalOutlierFactor(n_neighbors=20, contamination=contamination)
    df['lof_score'] = model.fit_predict(df[features])
    df['lof_anomaly'] = df['lof_score'].apply(lambda x: 1 if x == -1 else 0)

    anomaly_count = df['lof_anomaly'].sum()
    print(f"LOF detected {anomaly_count} anomalies ({anomaly_count/len(df)*100:.2f}%)")
    return df, model

def dbscan_detection(df, features):
    """
    Apply DBSCAN clustering for anomaly detection
    """
    print("\nRunning DBSCAN anomaly detection...")
    model = DBSCAN(eps=0.5, min_samples=5)
    df['dbscan_cluster'] = model.fit_predict(df[features])
    # In DBSCAN, points labeled as -1 are considered outliers
    df['dbscan_anomaly'] = df['dbscan_cluster'].apply(lambda x: 1 if x == -1 else 0)

    anomaly_count = df['dbscan_anomaly'].sum()
    print(f"DBSCAN detected {anomaly_count} anomalies ({anomaly_count/len(df)*100:.2f}%)")
    return df, model

def autoencoder_detection(df, features, threshold_percentile=95):
    """
    Apply Autoencoder neural network for anomaly detection
    """
    print("\nRunning Autoencoder anomaly detection...")
    # Split data for training the autoencoder
    X = df[features].values
    X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

    # Normalize data for autoencoder
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Build the autoencoder model
    input_dim = X_train_scaled.shape[1]
    encoding_dim = max(int(input_dim / 2), 2)  # Reduce dimension by half or minimum 2

    # Input layer
    input_layer = Input(shape=(input_dim,))

    # Encoder
    encoder = Dense(encoding_dim * 2, activation='relu')(input_layer)
    encoder = Dropout(0.2)(encoder)
    encoder = Dense(encoding_dim, activation='relu')(encoder)

    # Decoder
    decoder = Dense(encoding_dim * 2, activation='relu')(encoder)
    decoder = Dropout(0.2)(decoder)
    decoder = Dense(input_dim, activation='sigmoid')(decoder)

    # Autoencoder model
    autoencoder = Model(inputs=input_layer, outputs=decoder)
    autoencoder.compile(optimizer='adam', loss='mean_squared_error')

    # Train the autoencoder
    print("Training autoencoder...")
    autoencoder.fit(
        X_train_scaled, X_train_scaled,
        epochs=10,
        batch_size=32,
        shuffle=True,
        validation_data=(X_test_scaled, X_test_scaled),
        verbose=0
    )

    # Get reconstruction error
    X_all_scaled = scaler.transform(X)
    reconstructions = autoencoder.predict(X_all_scaled, verbose=0)
    mse = np.mean(np.square(X_all_scaled - reconstructions), axis=1)

    # Determine threshold based on percentile of reconstruction errors
    threshold = np.percentile(mse, threshold_percentile)
    print(f"Autoencoder reconstruction error threshold: {threshold:.6f}")

    df['autoencoder_score'] = mse
    df['autoencoder_anomaly'] = df['autoencoder_score'].apply(lambda x: 1 if x > threshold else 0)

    anomaly_count = df['autoencoder_anomaly'].sum()
    print(f"Autoencoder detected {anomaly_count} anomalies ({anomaly_count/len(df)*100:.2f}%)")
    return df, autoencoder

def ensemble_anomaly_detection(df):
    """
    Create an ensemble anomaly score by combining results from multiple detectors
    """
    # Check which anomaly detection methods were applied
    anomaly_columns = [col for col in df.columns if col.endswith('_anomaly')]

    if len(anomaly_columns) == 0:
        print("No anomaly detection methods found in the dataframe")
        return df

    # Create an ensemble score (sum of all individual anomaly flags)
    df['ensemble_score'] = df[anomaly_columns].sum(axis=1)

    # Flag as anomaly if at least half of the methods flagged it
    threshold = max(1, len(anomaly_columns) // 2)
    df['ensemble_anomaly'] = df['ensemble_score'].apply(lambda x: 1 if x >= threshold else 0)

    anomaly_count = df['ensemble_anomaly'].sum()
    print(f"\nEnsemble method detected {anomaly_count} anomalies ({anomaly_count/len(df)*100:.2f}%)")
    return df


# 3. Visualization Functions

def visualize_features_distribution(df, numeric_cols):
    """
    Visualize the distribution of key numeric features
    """
    print("\nVisualizing feature distributions...")
    # Select a subset of features for visualization if there are many
    if len(numeric_cols) > 6:
        selected_features = numeric_cols[:6]
    else:
        selected_features = numeric_cols

    fig, axes = plt.subplots(len(selected_features), 1, figsize=(10, 3*len(selected_features)))
    if len(selected_features) == 1:
        axes = [axes]

    for i, feature in enumerate(selected_features):
        sns.histplot(df[feature], ax=axes[i], kde=True)
        axes[i].set_title(f'Distribution of {feature}')

    plt.tight_layout()
    plt.show()

def visualize_pca_with_anomalies(df, features, anomaly_column):
    """
    Visualize data points in 2D PCA space with anomalies highlighted
    """
    print(f"\nVisualizing PCA projection with {anomaly_column} anomalies...")
    # Apply PCA for dimensionality reduction
    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(df[features])

    # Create a dataframe with PCA results
    pca_df = pd.DataFrame({
        'PCA1': pca_result[:, 0],
        'PCA2': pca_result[:, 1],
        'anomaly': df[anomaly_column]
    })

    # Plot normal vs anomalous points
    plt.figure(figsize=(10, 8))
    sns.scatterplot(
        x='PCA1', y='PCA2',
        hue='anomaly',
        data=pca_df,
        palette={0: 'blue', 1: 'red'},
        alpha=0.6,
        s=50
    )
    plt.title(f'PCA projection with {anomaly_column}')
    plt.legend(['Normal', 'Anomaly'])
    plt.show()

    # Print explained variance
    print(f"PCA explained variance ratio: {pca.explained_variance_ratio_}")

def visualize_anomaly_correlation(df):
    """
    Visualize correlation between different anomaly detection methods
    """
    anomaly_columns = [col for col in df.columns if col.endswith('_anomaly')]

    if len(anomaly_columns) <= 1:
        print("Not enough anomaly detection methods for correlation analysis")
        return

    print("\nVisualizing correlation between anomaly detection methods...")
    correlation = df[anomaly_columns].corr()

    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
    plt.title('Correlation between different anomaly detection methods')
    plt.show()

def visualize_feature_importance(df, features, anomaly_column):
    """
    Visualize feature importance for distinguishing anomalies
    """
    print(f"\nAnalyzing feature importance for {anomaly_column}...")

    # Calculate mean difference for each feature between normal and anomalous points
    normal_df = df[df[anomaly_column] == 0]
    anomaly_df = df[df[anomaly_column] == 1]

    # Initialize empty dictionaries to store metrics
    mean_diff = {}
    std_ratio = {}

    for feature in features:
        # Calculate mean difference (how far apart are the means)
        mean_normal = normal_df[feature].mean()
        mean_anomaly = anomaly_df[feature].mean()
        mean_diff[feature] = abs(mean_normal - mean_anomaly)

        # Calculate standard deviation ratio (how much more variable is one group)
        std_normal = normal_df[feature].std()
        std_anomaly = anomaly_df[feature].std()
        if std_normal > 0 and std_anomaly > 0:
            std_ratio[feature] = max(std_normal, std_anomaly) / min(std_normal, std_anomaly)
        else:
            std_ratio[feature] = 1.0

    # Create DataFrames for visualization
    mean_diff_df = pd.DataFrame({'Feature': list(mean_diff.keys()),
                                'Mean Difference': list(mean_diff.values())})
    mean_diff_df = mean_diff_df.sort_values('Mean Difference', ascending=False)

    std_ratio_df = pd.DataFrame({'Feature': list(std_ratio.keys()),
                                'Std Ratio': list(std_ratio.values())})
    std_ratio_df = std_ratio_df.sort_values('Std Ratio', ascending=False)

    # Plot feature importance based on mean difference
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Mean Difference', y='Feature', data=mean_diff_df.head(10))
    plt.title(f'Top Features by Mean Difference for {anomaly_column}')
    plt.tight_layout()
    plt.show()

def visualize_anomaly_timeline(df):
    """
    Visualize anomalies over time
    """
    if 'timestamp' not in df.columns:
        print("Timestamp column not found, cannot create timeline visualization")
        return

    print("\nVisualizing anomalies over time...")

    # Check for timestamp column and convert if needed
    if df['timestamp'].dtype != 'datetime64[ns]':
        df['timestamp'] = pd.to_datetime(df['timestamp'])

    # Resample data by hour and count anomalies
    anomaly_columns = [col for col in df.columns if col.endswith('_anomaly')]

    plt.figure(figsize=(12, 6))

    # Group by hour and count anomalies
    for anomaly_col in anomaly_columns:
        if anomaly_col != 'ensemble_anomaly':  # Skip ensemble for clarity if there are many methods
            hourly_anomalies = df.set_index('timestamp').resample('H')[anomaly_col].sum()
            hourly_anomalies.plot(label=anomaly_col)

    plt.title('Anomaly Counts Over Time')
    plt.xlabel('Time')
    plt.ylabel('Number of Anomalies')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


# 4. Main Execution Function

def detect_network_anomalies(file_path, contamination=0.05):
    """
    Main function to run the entire anomaly detection pipeline
    """
    print("Starting network anomaly detection pipeline...\n")

    # Step 1: Load and preprocess data
    df = load_network_data(file_path)
    df_processed, features, scaler = preprocess_data(df)

    # Step 2: Visualize feature distributions
    visualize_features_distribution(df_processed, features)

    # Step 3: Apply multiple anomaly detection algorithms
    # 3.1: Isolation Forest
    df_processed, iforest_model = isolation_forest_detection(df_processed, features, contamination)

    # 3.2: Local Outlier Factor
    df_processed, lof_model = lof_detection(df_processed, features, contamination)

    # 3.3: DBSCAN Clustering
    df_processed, dbscan_model = dbscan_detection(df_processed, features)

    # 3.4: Autoencoder
    df_processed, autoencoder_model = autoencoder_detection(df_processed, features)

    # Step 4: Create ensemble anomaly score
    df_processed = ensemble_anomaly_detection(df_processed)

    # Step 5: Visualize results
    # 5.1: PCA visualization with anomalies
    visualize_pca_with_anomalies(df_processed, features, 'ensemble_anomaly')

    # 5.2: Correlation between different anomaly detection methods
    visualize_anomaly_correlation(df_processed)

    # 5.3: Feature importance for anomaly detection
    visualize_feature_importance(df_processed, features, 'ensemble_anomaly')

    # 5.4: Timeline of anomalies
    visualize_anomaly_timeline(df_processed)

    # Step 6: Return processed data with anomaly scores
    print("\nAnomaly detection complete!")

    # Return a summary of detected anomalies
    anomaly_columns = [col for col in df_processed.columns if col.endswith('_anomaly')]
    anomaly_summary = df_processed[anomaly_columns].sum().reset_index()
    anomaly_summary.columns = ['Method', 'Count']
    anomaly_summary['Percentage'] = (anomaly_summary['Count'] / len(df_processed) * 100).round(2)

    print("\nAnomaly Detection Summary:")
    print(anomaly_summary)

    return df_processed


# 5. Example Usage

if __name__ == "__main__":
    # Replace with actual file path or use sample data
    file_path = "network_traffic.csv"  # This will trigger the sample data generation if file not found

    # Run the anomaly detection pipeline
    results = detect_network_anomalies(file_path, contamination=0.05)

    # Export results to CSV (optional)
    results.to_csv("network_traffic_anomalies.csv", index=False)

    # Display sample of anomalies detected by ensemble method
    print("\nSample of detected anomalies:")
    anomalies = results[results['ensemble_anomaly'] == 1].head(10)
    print(anomalies[['bytes_sent', 'bytes_received', 'duration', 'packet_count',
                    'isolation_forest_anomaly', 'lof_anomaly', 'dbscan_anomaly',
                    'autoencoder_anomaly', 'ensemble_anomaly']])
```
